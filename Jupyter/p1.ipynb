```python
limit_duplicated=2000
limit_unique=100

for col in all_cols:
    #print(type(df[col][0]))
    col_unique_values=0
    #print(pd.unique(df[col].values))
    try:
        #print(len(df[col].unique()))
        col_unique_values=len(df[col].unique())
    except:
        temp_list=[]
        for sublist in df[col]:
            #print(type(sublist))
            if type(sublist)==list:
                for item in sublist:
                    temp_list.append(item)
            else:
                temp_list.append(sublist)
        #print(temp_list)
        temp_set = set(temp_list)
        #print(len(temp_set))
        col_unique_values=len(temp_set)
    #print(len(str(df[col].to_string().unique())))
    print("unique",col_unique_values)
    if col_unique_values > limit_unique:
        #print(df[df[col].duplicated()])
        print(df[df[col].duplicated()].shape)
        #if df.iloc[:,1].isnull().sum()
        if df.loc[:,col].isnull().sum() >1000:
            if df[df[col].duplicated()].shape[0]>limit_duplicated:
                df.drop(col, axis='columns',inplace=True)
    
    #print(df[df[col].duplicated()])
    #print(df[df[col].duplicated()].shape)
    #if df[df[col].duplicated()].shape[0]>limit:
        #df.drop(col, axis='columns',inplace=True)
```
